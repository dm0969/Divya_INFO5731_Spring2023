## (5) (10 extra points) Compare the results generated by the four topic modeling algorithms, which one is better? You should explain the reasons in details.

Similar documents appear in the top and bottom categories of all three algorithms' lists of 100 patent papers. Because it has the capability to decompose the original matrix using an information retrieval technique while maintaining key topics, LSA produces better results than LTA and BERT. LDA operates under the presumption that documents are produced from a variety of subjects, and that words are then output using a probability distribution. The context/paragraph vectors of LDA2Vec have the disadvantage of resembling standard word vectors, which makes them less interpretable as, say, the result of LDA. By working with text fragments the size of documents and breaking down the document vectors into two distinct components, the lda2vec model moves one step beyond the paragraph vector approach. A document vector is divided into a subject matrix and a document weight vector, much like the LDA model. The topic matrix is made up of the various topic vectors, whereas the document weight vector shows the percentage of the various subjects. Thus, a context vector is created by fusing the various subject vectors that appear in a document.
